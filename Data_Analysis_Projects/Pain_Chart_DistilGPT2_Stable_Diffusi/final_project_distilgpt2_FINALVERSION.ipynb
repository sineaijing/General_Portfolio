```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained('distilgpt2',do_lower_case=False)

def tokenize_function(examples):
  return tokenizer(examples['text'],truncation=True)

tokenized_data = data.map(tokenize_function,batched=True,num_proc=4,batch_size=1)

tokenized_data.set_format('torch',columns=['input_ids','attention_mask'])

#tokenized_data_train = data_train.map(tokenize_function,batched=True)
#tokenized_data_test = data_test.map(tokenize_function,batched=True)
```

```python
# use float16 instead of float32 to avoid memory over usage      
from torch.cuda.amp import autocast, GradScaler

# Create a custom Trainer class with mixed-precision training
class MyTrainer(Trainer):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.scaler = GradScaler()

    def compute_loss(self, model, inputs, return_outputs=False):
        # Convert inputs to float16 and scale the loss
        inputs = {k: v.half() if isinstance(v, torch.Tensor) else v for k, v in inputs.items()}
        with autocast():
            outputs = model(**inputs)
            loss = outputs.loss
        scaled_loss = self.scaler.scale(loss)

        # Return scaled loss or outputs
        if return_outputs:
            return scaled_loss, outputs
        else:
            return scaled_loss

# create dataloader to not import all data at once
from torch.utils.data import DataLoader, IterableDataset

class MyDataset(IterableDataset):
    def __init__(self, filepath):
        self.filepath = filepath
    
    def __iter__(self):
        def file_reader(filepath):
            with open(filepath, 'r') as f:
                for line in f:
                    yield line.strip()
        
        for line in file_reader(self.filepath):
            yield torch.tensor(line.split(), dtype=torch.float16)
        
train_dataset = tokenized_data['train']
train_loader = DataLoader(train_dataset, batch_size=1)

for batch in train_loader:
    # Do your training here
    pass
```

```python
# define hyperparameters in trainer
args = TrainingArguments(
    output_dir='model_finetuned_data',       # save the finetuned model
    num_train_epochs=10,
    evaluation_strategy='epoch',
    save_strategy = 'epoch',
    learning_rate=1e-5,
    gradient_accumulation_steps=2,
    weight_decay=0.01,
    fp16=True,      # enables mized precision training
    per_device_train_batch_size = 1,     # wowooowooow this line of code FIXED it! 
    #dropout_rate=0.2,           # chances of dropping out a neuron 
    #save_total_limit=2,
    #logging_dir='./logs',
    #logging_steps=500,
    #evaluation_strategy='steps',
    eval_steps=500,
    load_best_model_at_end=True,
    adam_epsilon=1e-5,
    #callbacks=[clear_cache_callback]
    )
```

```python
from transformers import EarlyStoppingCallback

# add early stopping to prevent overfitting 
callbacks = [EarlyStoppingCallback(early_stopping_patience=5)]

# trainer object! 
trainer = Trainer(
    model=model,
    args = args,
    #train_dataset = small_train_dataset,
    #eval_dataset = small_eval_dataset,
    train_dataset = tokenized_data['train'],
    eval_dataset = tokenized_data['test'],
    #compute_metrics = compute_metrics,
    data_collator = data_collator,
    callbacks = callbacks
    )
```
